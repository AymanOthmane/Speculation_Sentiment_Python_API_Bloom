```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Sample data fetching step
data = df_SSI_BBG['2009-12-31':].copy()

data.rename(columns={'SSI': 'Close'}, inplace=True)
symbol = 'SSI'

# Normalize the data
data['Normalized_Close'] = (data['Close'] - data['Close'].mean()) / data['Close'].std()

# Create features
data['Returns'] = data['Normalized_Close'].pct_change()
data['SMA'] = data['Normalized_Close'].rolling(window=20).mean()
data['StdDev'] = data['Normalized_Close'].rolling(window=20).std()
data = data.dropna()

# Ornstein-Uhlenbeck Process Estimation
def estimate_ou_params(X):
    dt = 1  # assuming daily data

    def negative_log_likelihood(params):
        theta, mu, sigma = params
        X_diff = np.diff(X)
        X_lag = X[:-1]
        residuals = X_diff - theta * (mu - X_lag) * dt
        sigma = max(sigma, 1e-6)  # Prevent sigma from being zero
        # log_likelihood = -np.sum(-0.5 * np.log(2 * np.pi * sigma**2 * dt) - residuals**2 / (2 * sigma**2 * dt))
        log_likelihood = -(0.5 * len(residuals) * (np.log(2*np.pi))) - 0.5 * np.sum(np.log(sigma**2)) - 0.5 * np.sum((residuals/sigma)**2)
        # Regularization term to prevent extreme parameter values
        regularization = 0.01 * (theta**2 + mu**2 + sigma**2)
        return -log_likelihood + regularization

    initial_params = [0.1, np.mean(X), np.std(X)]
    bounds = [(0.001, 5), (-np.max(X), np.max(X)), (1e-6, np.std(X) * 3)]
    result = minimize(negative_log_likelihood, initial_params, bounds=bounds, method='L-BFGS-B')
    
    if result.success:
        theta, mu, sigma = result.x
    else:
        raise ValueError("Optimization failed")

    return theta, mu, sigma

# Estimate OU parameters
theta, mu, sigma = estimate_ou_params(data['Normalized_Close'].values)

# Calculate Residuals
data['Residual'] = data['Normalized_Close'] - mu

# Identify Significant Turning Points
std_residual = data['Residual'].rolling(window=20).std()
threshold = 2 * std_residual  # Adjust the multiplier based on your needs
data['TurningPoint'] = np.where(np.abs(data['Residual'].diff()) > threshold, 1, 0)

# Remove the first few rows with NaN values
data = data.dropna()

# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(data['Normalized_Close'], label='Normalized SSI', color='blue')
turning_points = data[data['TurningPoint'] != 0]
plt.scatter(turning_points.index, turning_points['Normalized_Close'], c='red', label='Turning Points', marker='o')
plt.title(f'{symbol} - Predicted Turning Points with Ornstein-Uhlenbeck Process')
plt.legend()
plt.show()

# Display the estimated parameters
print(f"Estimated parameters: theta={theta}, mu={mu:}, sigma={sigma}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from imblearn.over_sampling import SMOTE
from ta import add_all_ta_features
from collections import Counter

# Load and preprocess data
data = df_SSI_BBG.copy()
data.rename(columns={'SSI': 'Close'}, inplace=True)
data.replace(0, 1e-6, inplace=True)
data = data.dropna()
symbol = 'SSI'

# Feature Engineering
data['Returns'] = data['Close'].pct_change()
data['SMA'] = data['Close'].rolling(window=20).mean()
data['StdDev'] = data['Close'].rolling(window=20).std()
data = add_all_ta_features(data, open="Close", high="Close", low="Close", close="Close", volume="Close", fillna=True)
data['TurningPoint'] = (np.sign(data['Returns'].shift(1)) != np.sign(data['Returns'])).astype(int)
data = data.dropna()

# Define features and target
features = [col for col in data.columns if col not in ['Close', 'TurningPoint']]
X = data[features]
y = data['TurningPoint']

# Scale the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Prepare the dataset for LSTM
def create_dataset(X, y, time_step=1):
    Xs, ys = [], []
    for i in range(len(X) - time_step):
        Xs.append(X[i:(i + time_step)])
        ys.append(y[i + time_step])
    return np.array(Xs), np.array(ys)

time_step = 20
X, y = create_dataset(X_scaled, y.values, time_step)

# Split the data into training and testing sets
train_size = int(len(X) * 0.7)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Flatten the data for SMOTE
n_samples, time_steps, n_features = X_train.shape
X_train_flat = X_train.reshape(-1, n_features)
y_train_flat = np.repeat(y_train, time_steps)

# Apply SMOTE
sm = SMOTE(random_state=42)
X_train_res_flat, y_train_res_flat = sm.fit_resample(X_train_flat, y_train_flat)

# Reshape back to 3D
X_train_res = X_train_res_flat.reshape(-1, time_steps, n_features)
y_train_res = y_train_res_flat[::time_steps]

print(f'Class distribution after resampling: {Counter(y_train_res)}')

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(time_step, X_train_res.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train_res, y_train_res, validation_data=(X_test, y_test), epochs=100, batch_size=32, callbacks=[early_stop], verbose=1)

# Predict turning points
y_pred_prob = model.predict(X_test)

# Experiment with different thresholds
threshold = 0.5  # You can adjust this value
y_pred = (y_pred_prob > threshold).astype(int)

# Evaluate the model
def evaluate_model(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, zero_division=1)
    precision = precision_score(y_true, y_pred, zero_division=1)
    recall = recall_score(y_true, y_pred)
    print(f'{model_name} - Accuracy: {accuracy:.2f}, F1 Score: {f1:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}')
    print(classification_report(y_true, y_pred, zero_division=1))

evaluate_model(y_test, y_pred, 'LSTM Model')

# Plot the results
data['Predicted_TurningPoint'] = 0
data.iloc[-len(y_pred):, data.columns.get_loc('Predicted_TurningPoint')] = y_pred.flatten()

plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='SSI', color='blue')
turning_points = data[data['Predicted_TurningPoint'] != 0]
plt.scatter(turning_points.index, turning_points['Close'], c='red', label='Predicted Turning Points', marker='o')
plt.title(f'{symbol} - Predicted Turning Points with LSTM')
plt.legend()
plt.show()
```
